```python
# Q.7 Solution
def vowels(string):
  vowels = "aeiouAEIOU"
  count = 0
  for i in string:
    if i in vowels:
      count += 1
  print(count)
string = input("Enter a String:")
vowels(string)
```

```python
# Q.8 Solution
def second_largest(list_1):
  sorted_list = sorted(set(list_1))
  print(f" {sorted_list[-2]}")
list_1 = [ 52,64,68,59,69 ]
second_largest(list_1)
```
